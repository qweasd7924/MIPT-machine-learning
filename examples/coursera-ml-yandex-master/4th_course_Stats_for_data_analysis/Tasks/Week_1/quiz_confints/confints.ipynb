```python
import pandas as pd
import numpy as np
from statsmodels.stats.weightstats import _zconfint_generic, _tconfint_generic
```

```python
data = pd.read_csv('water.csv', delimiter='\t')
```

```python
data
```

```python
mortality_data = data['mortality']
```

```python
mortality_mean = mortality_data.mean()
```

```python
mortality_mean_std = mortality_data.std(ddof=1)/np.sqrt(len(mortality_data))
```

```python
print("mortality 95%% confidence interval {}".format(_tconfint_generic(mortality_mean, mortality_mean_std,
                                                                       len(mortality_data) - 1,
                                                                       0.05, 'two-sided')))
```

```python
mortality_south_data = data[data['location'] == 'South']['mortality']
```

```python
mortality_south_mean = mortality_south_data.mean()
```

```python
mortality_south_mean_std = mortality_south_data.std(ddof=1)/np.sqrt(len(mortality_south_data))
```

```python
print("South cities mortality 95%% confidence interval {}".format(_tconfint_generic(mortality_south_mean, 
                                                                                    mortality_south_mean_std,
                                                                                    len(mortality_south_data) - 1,
                                                                                    0.05, 'two-sided')))
```

```python
mortality_north_data = data[data['location'] == 'North']['mortality']
```

```python
mortality_north_mean = mortality_north_data.mean()
```

```python
mortality_north_mean_std = mortality_north_data.std(ddof=1)/np.sqrt(len(mortality_north_data))
```

```python
print("North cities mortality 95%% confidence interval {}".format(_tconfint_generic(mortality_north_mean, 
                                                                                    mortality_north_mean_std,
                                                                                    len(mortality_north_data) - 1,
                                                                                    0.05, 'two-sided')))
```

```python
hardness_south_data = data[data['location'] == 'South']['hardness']
hardness_north_data = data[data['location'] == 'North']['hardness']

hardness_south_mean = hardness_south_data.mean()
hardness_north_mean = hardness_north_data.mean()

hardness_south_mean_std = hardness_south_data.std(ddof=1)/np.sqrt(len(hardness_south_data))
hardness_north_mean_std = hardness_north_data.std(ddof=1)/np.sqrt(len(hardness_north_data))

print("South cities hardness 95%% confidence interval {}".format(_tconfint_generic(hardness_south_mean, 
                                                                                    hardness_south_mean_std,
                                                                                    len(hardness_south_data) - 1,
                                                                                    0.05, 'two-sided')))

print("North cities hardness 95%% confidence interval {}".format(_tconfint_generic(hardness_north_mean, 
                                                                                    hardness_north_mean_std,
                                                                                    len(hardness_north_data) - 1,
                                                                                    0.05, 'two-sided')))
```
