Один из эффективных методов первичного анализа данных - отображение попарных зависимостей признаков. Создается $m \times m$ графиков (*m* - число признаков), где по диагонали рисуются гистограммы распределения признаков, а вне диагонали - scatter plots зависимости двух признаков. Это можно делать с помощью метода $scatter\_matrix$ Pandas Data Frame или *pairplot* библиотеки Seaborn. 

Чтобы проиллюстрировать этот метод, интересней добавить третий признак. Создадим признак *Индекс массы тела* ([BMI](https://en.wikipedia.org/wiki/Body_mass_index)). Для этого воспользуемся удобной связкой метода *apply* Pandas DataFrame и lambda-функций Python.

```python
def make_bmi(height_inch, weight_pound):
    METER_TO_INCH, KILO_TO_POUND = 39.37, 2.20462
    return (weight_pound / KILO_TO_POUND) / \
           (height_inch / METER_TO_INCH) ** 2
```

```python
data['BMI'] = data.apply(lambda row: make_bmi(row['Height'], 
                                              row['Weight']), axis=1)
```

## Задание 2. Минимизация квадратичной ошибки

В простейшей постановке задача прогноза значения вещественного признака по прочим признакам (задача восстановления регрессии) решается минимизацией квадратичной функции ошибки. 

**[6].** Напишите функцию, которая по двум параметрам $w_0$ и $w_1$ вычисляет квадратичную ошибку приближения зависимости роста $y$ от веса $x$ прямой линией $y = w_0 + w_1 * x$:
$$error(w_0, w_1) = \sum_{i=1}^n {(y_i - (w_0 + w_1 * x_i))}^2 $$
Здесь $n$ – число наблюдений в наборе данных, $y_i$ и $x_i$ – рост и вес $i$-ого человека в наборе данных. 

```python
# Ваш код здесь
# Определяем функцию ошибки
# w_vec - вектор весов: w_vec[0] - w0, w_vec[1] - w1 
def Q(w_vec):
    return sum((data['Height'] - (w_vec[0] + w_vec[1]*data['Weight']))**2)    
```

Итак, мы решаем задачу: как через облако точек, соответсвующих наблюдениям в нашем наборе данных, в пространстве признаков "Рост" и "Вес" провести прямую линию так, чтобы минимизировать функционал из п. 6. Для начала давайте отобразим хоть какие-то прямые и убедимся, что они плохо передают зависимость роста от веса.

**[7].** Проведите на графике из п. 5 Задания 1 две прямые, соответствующие значениям параметров ($w_0, w_1) = (60, 0.05)$ и ($w_0, w_1) = (50, 0.16)$. Используйте метод *plot* из *matplotlib.pyplot*, а также метод *linspace* библиотеки NumPy. Подпишите оси и график.

[Документация  matplotlib.pyplot.plot](http://matplotlib.org/api/pyplot_api.html#matplotlib.pyplot.plot)

```python
# Ваш код здесь
# Задаем вектор значений параметров (w0, w1) = (60, 0.05)
w_vec1 = [60, 0.05]       
# Задаем вектор значений параметров (w0, w1) = (50, 0.16)
w_vec2 = [50, 0.16]

# формируем график зависимости роста (ось у) от веса (ось x)
data.plot(x='Weight', 
          y='Height',
          kind='scatter')

# Подписываем оси графика: ось x – "Вес", ось y - "Рост" 
plt.xlabel(u'Вес')
plt.ylabel(u'Рост')

#задаём линейную модель для прогнозирования роста подростка по его весу
fy = (lambda x,w_vec: w_vec[0] + w_vec[1]*x)

#генерируем два произвольных значения параметра "Вес" - Weight 
xn = np.linspace(min(data['Weight']), max(data['Weight']), 2)

#опеделяем переменную yn - массив ответов линейной модели fy
yn = np.ndarray(len(xn))

#вычисляем прогнозные значения роста (Height), с параметрами W0=60 и W1=0.05
yn = fy(xn, w_vec1) 
plt.plot(xn, yn, 'r') #рисуем красную линию - 'r'

#вычисляем прогнозные значения роста (Height), с параметрами W0=50 и W1=0.16
yn = fy(xn, w_vec2)
plt.plot(xn, yn, 'g') #рисуем зелёную линию - 'g'
```

Минимизация квадратичной функции ошибки - относительная простая задача, поскольку функция выпуклая. Для такой задачи существует много методов оптимизации. Посмотрим, как функция ошибки зависит от одного параметра (наклон прямой), если второй параметр (свободный член) зафиксировать.

**[8].** Постройте график зависимости функции ошибки, посчитанной в п. 6, от параметра $w_1$ при $w_0$ = 50. Подпишите оси и график.

```python
# Ваш код здесь
#создаем массив NumPy c множеством значений параметра w1 в диапозоне [-5,5]
wn1 = np.arange(-5, 5, 0.25) 

#определяем список значений функции ошибки Q([w0, w1])
Qval = list()

#фиксируем значение второго параметра (свободный член)
w0 = 50  

for w1 in wn1: #для всех значений первого параметра - w1 
    Qval.append(Q([w0, w1]))  #вычисляем значение функции ошибки Q([w0 = 50, w1])

#строим график зависимости зачений Q(w0,w1) от w1, при w0 = 50
plt.plot(wn1, Qval, 'r') #рисуем красную линию - 'r'

# Подписываем: ось y - "Q(w0, w1)", ось x – "w1" 
plt.xlabel('w1')
plt.ylabel('Q(w0,w1)')
```

Теперь методом оптимизации найдем "оптимальный" наклон прямой, приближающей зависимость роста от веса, при фиксированном коэффициенте $w_0 = 50$.

**[9].** С помощью метода *minimize_scalar* из *scipy.optimize* найдите минимум функции, определенной в п. 6, для значений параметра  $w_1$ в диапазоне [-5,5]. Проведите на графике из п. 5 Задания 1 прямую, соответствующую значениям параметров ($w_0$, $w_1$) = (50, $w_1\_opt$), где $w_1\_opt$ – найденное в п. 8 оптимальное значение параметра $w_1$. 

[Документация minimize_scalar](http://docs.scipy.org/doc/scipy/reference/generated/scipy.optimize.minimize_scalar.html)

```python
# Ваш код здесь
# Импортируем модуль optimize библиотеки scipy
from scipy import optimize 

#фиксируем значение второго параметра (свободный член)
w0 = 50

#определяем лямбда-функцию с одним параметром - w1, при фиксированном w0
Q_w0_fix = (lambda w1: Q([w0,w1]))

# находим локальный минимум функции Q_lambda(w1) на отрезке [-5, 5]
w1_opt = optimize.minimize_scalar(Q_w0_fix,
                                  bounds= (-5, 5),
                                  method='bounded')
print(w1_opt)
```

```python
# Ваш код здесь

# формируем график зависимости роста (ось у) от веса (ось x)
data.plot(x='Weight', 
          y='Height',
          kind='scatter')

# Подписываем: ось x – "Вес", ось y - "Рост" 
plt.xlabel(u'Вес')
plt.ylabel(u'Рост')

#задаём линейную модель для прогнозирования роста подростка по его весу
#с учётом найденого оптимального значения w1_opt, при w0 = 50
w0 = 50
w1 = w1_opt.x
f_opt = (lambda x: w0 + w1*x)

#генерируем два произвольных значения параметра "Вес" - Weight 
xn = np.linspace(min(data['Weight']), max(data['Weight']),2)

# вычисляем прогнозные значения роста (Height), с параметрами W1=w1_opt и W0=50 (по умолчанию)
yn = f_opt(xn)
plt.plot(xn, yn, 'y') #рисуем жёлтую линию - 'y'
```

Посмотрим, как в Python рисовать 3D картинки, на примере отображения функции $z(x,y) = sin(\sqrt{x^2+y^2})$ для значений $x$ и $y$ из интервала [-5,5] c шагом 0.25.

**[10].** Постройте 3D-график зависимости функции ошибки, посчитанной в п.6 от параметров $w_0$ и $w_1$. Подпишите ось $x$ меткой «Intercept», ось $y$ – меткой «Slope», a ось $z$ – меткой «Error».

[Документация plot_surface](http://matplotlib.org/mpl_toolkits/mplot3d/tutorial.html)

```python
fig = plt.figure()
ax = fig.gca(projection='3d') # get current axis

# Создаем массивы NumPy с координатами точек по осям X и У. 
Wn0 = np.arange(-5, 5, 0.25) #массив значений параметра w0 (ось X) - Intercept
Wn1 = np.arange(-5, 5, 0.25) #массив значений параметра w1 (ось Y) - Slope

# Используем метод meshgrid, при котором по векторам координат 
# создается матрица координат.
Wn0, Wn1 = np.meshgrid(Wn1, Wn0)

# Формируем массивы значений роста и веса для вычисления функции ошибки Qval
Yn  = data['Height'].head(len(Wn1)).values #возращает N первых элементов в виде numpy.ndarray
Xn  = data['Weight'].head(len(Wn1)).values #возращает N первых элементов в виде numpy.ndarray

#Вычисляем значение функции ошибки 
Qval = (Yn - (Wn0 + Wn1*Xn))**2

# Наконец, используем метод *plot_surface* объекта типа Axes3DSubplot. 
# Значения параметра color - http://matplotlib.org/examples/color/named_colors.html
surf = ax.plot_surface(Wn0, Wn1, Qval, color = 'c')

# Подптисываем оси
ax.set_xlabel('Intercept')
ax.set_ylabel('Slope')
ax.set_zlabel('Error')
plt.show()
```

```python
w_vec_opt = optimize.minimize(Q,                                 # функция качества Q(w_vec)
                              x0 = (0,0),                        # начальные условия (w0,w1) = (0, 0)
                              method='L-BFGS-B',                 # метод оптимизации 'L-BFGS-B' 
                              bounds = ((-100, 100), (-5, 5)))   # ограничения диапозона для w0 = [-100,100] и w1 = [-5, 5]

print(w_vec_opt)
```

```python
# формируем график зависимости роста (ось у) от веса (ось x)
data.plot(x='Weight', 
          y='Height',
          kind='scatter')

# Подписываем: ось x – "Вес", ось y - "Рост" 
plt.xlabel(u'Вес')
plt.ylabel(u'Рост')

#задаём линейную модель для прогнозирования роста подростка по его весу
#с учётом найденого минимума функции Q - w1_vec_opt.x
w0 = w_vec_opt.x[0] # значение параметра w0
w1 = w_vec_opt.x[1] # значение параметра w1
f_opt = (lambda x: w0 + w1*x)

#генерируем два произвольных значения параметра "Вес" - Weight 
xn = np.linspace(min(data['Weight']), max(data['Weight']), 2)

# вычисляем прогнозные значения роста (Height), с оптимальными параметрами w0 и w1 
yn = f_opt(xn)

plt.plot(xn, yn, 'gold') #рисуем золотую линию - 'gold'
```
