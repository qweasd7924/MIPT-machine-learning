# Распределение Рэлея

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
import math

%matplotlib inline
```

```python
x = np.linspace(0,15,100)
al = 4 # переменная распределения
dispAl = (2 - math.pi/2) * al ** 2 #диспеерсия
moAl = math.sqrt(math.pi/2) * al #мат ожидание
rv = sts.rayleigh(0,al)
cdf = rv.cdf(x)
plt.plot(x, cdf, label="$al=%s$" % al)
plt.legend()
plt.title("CDF rayleigh")
```

```python
x = np.linspace(0,15,100)
rv = sts.rayleigh(0, al)
pdf = rv.pdf(x)
plt.plot(x, pdf, label="$al=%s$" % al)
plt.legend()
plt.title("PDF rayleigh")
```

# n=5

```python
n = 5
```

```python
sample = rv.rvs(n)
```

```python
plt.hist(sample, bins = n, normed=True)
plt.ylabel('number of samples')
plt.xlabel('$x$')
```

```python
i = 0
vybSredList = []
while i < 1000:
    sample = rv.rvs(n)
    vybSredList.append(sum([t for t in sample])/n)
    i += 1
plt.hist(vybSredList, bins = 50, normed=True)
plt.ylabel('number of samples')
plt.xlabel('$x$')
```

```python
norm = sts.norm(moAl, math.sqrt(dispAl/n))

plt.hist(vybSredList, bins = 50, normed=True)
plt.ylabel('number of samples')
plt.xlabel('$x$')

x = np.linspace(1,9,100)
pdf = norm.pdf(x)
plt.plot(x, pdf, label='theoretical pdf', alpha=0.5)
```
