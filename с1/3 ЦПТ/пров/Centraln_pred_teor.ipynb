## Распределение Рэлея

Сгенерируйте из выбранного непрерывного распределения выборку объёма 1000, постройте гистограмму выборки и нарисуйте поверх неё теоретическую плотность распределения вашей случайной величины.

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
from scipy.stats import rayleigh
from scipy.stats import norm
```

```python
r = rayleigh.rvs(size=1000) #Выборка из 1000 со стандартными параметрами
x = np.linspace(rayleigh.ppf(0.001),rayleigh.ppf(0.99), 100) #https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.rv_continuous.ppf.html
#Работа с графиком
plt.hist(r,  normed=True, label='Выборка из 1000')
plt.plot(x, rayleigh.pdf(x), label='Теоретическая плотность' )
plt.title('Гистограмма для выборки объема 1000 и её теоретическая плотность')
plt.ylabel('fraction of samples')
plt.xlabel('$x$')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.show()
```

## Распределение выборочного среднего

### Математическое ожидание

Здесь сигма - scale parameter равен по умолчанию единице

### Дисперсия

Здесь сигма - scale parameter равен по умолчанию единице

```python
def sample_gen(n): return([rayleigh.rvs(size=n).mean()  for k in range(1000)])# функция 1000 значений матожидания для выборки n 
```

### Размер выборки n = 5


```python
n=5
x = np.linspace(0.5,2,1000)

sample=sample_gen(n)
plt.hist(sample, bins=20, density=True, label='n= ' + str(n))

mean, variance = rayleigh.stats(moments='mv') #матожидание и дисперсия
norm_rv = sts.norm(mean, ((variance/n)**0.5))#Согласно ЦПТ дисперсия равна дисперсии изначального распределения делить на n
plt.plot(x, norm_rv.pdf(x), label='Теоретическое нормальное для n= ' + str(n))
plt.title('Гистограмма для распределение выборочного среднего из '+ str(n) +' и соответсвующее нормальное')
plt.ylabel('fraction of samples')
plt.xlabel('$x$')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.show()
```
