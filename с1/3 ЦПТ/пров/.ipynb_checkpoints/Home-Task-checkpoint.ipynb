```python
import pandas as pd
import numpy as np 
import scipy.stats as sts
import matplotlib.pyplot as plt
import math
%matplotlib inline
```

#  Определяем распределение Лаплас
Про распределение Лаплас можно почитать [здесь](https://ru.wikipedia.org/wiki/Распределение_Лапласа)

Мат ожидание: E(x) = β, β - значение который мы разбили [-∞; ∞] на [-∞; β] и [β; ∞]

Дисперсия: D(x) = 2/α**2

где 

α >0 — параметр масштаба, 

-∞<β<+∞ — параметр сдвига.


```python
beta = 2 # равен на средний beta = mu
alpha = 2
disp = 2/alpha**2  # дисперсия
```

```python
mu = 2.0 # отвечает за положение графика на числовой оси 
delta = 0.5  # отвечает за форму в графике

laplace_rv = sts.laplace(mu, delta) #задаем функцию
sample = laplace_rv.rvs(1000) #генерим случайную выборку 1000 значений
x = np.linspace(0, 5, 1000) #определяем массив точек для построения плотности распределения
pdf = laplace_rv.pdf(x) #вычисляем значения плотности распределения от выборки

#строим гистрограмму выборки и теоретическую функцию распределения
plt.plot(x, pdf, label = 'theoritical PDF')
plt.hist(sample, bins = 60, range = (0, 10), normed = True)
plt.legend()
plt.ylabel('$f(x)$/number of samples')
plt.xlabel('$x$')
plt.show()
```

# Оценим распределение выборочного среднего случайной величины при разных объёмах выборок¶

## при n = 5, n = 40, n = 100


```python
n_size = [5, 40, 100] # выборки n = 5, 40, 100

for size in n_size: 
    
    # Гистограмма выборки
    sample = [np.mean(laplace_rv.rvs(size=size)) for _ in range(size)]
    plt.hist(sample, normed=True, label = 'theoritical pdf')
    
    # теоретический плотность распределения
    # Сгенерируем выборку объёма из нормального распределения
    ex = 2 #среднее Лаплас
    dx = 0.5 #дисперсия Лаплас
    sigma = math.sqrt(dx/size) # sigma = sqrt(dx/size) - так определяется сигма на n выборке
    
    #Остальные как в видео которые мы смотрели, надеюсь, ничего сложного тут нету
    norm_rv = sts.norm(ex, sigma) 
    x = np.linspace(0, 5, 1000)
    pdf = norm_rv.pdf(x)
    
    plt.plot(x, pdf, label = 'PDF (n = %i)' % size)
    plt.ylabel("F(x)")
    plt.xlabel("x")
    plt.legend(loc = 'best')
    plt.show()
    
```
